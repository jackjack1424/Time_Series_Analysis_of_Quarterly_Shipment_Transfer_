## 4. Plot a quarterly comparison of the shipments using a barplot.

```python
df1=df
df1['Quarter']=df1['Quarter'].astype(str)
df1['Quarter_Year'] = df1['Quarter'] + ' ' + df1['Year'].astype(str)
```

## 5. Plot the year on year quarterly comparison of the shipments.

```python
import numpy as np
plt.figure(figsize=(10,8))
grouped_df = df1.groupby(["Year", "Quarter"])["Shipments"].sum().reset_index()
pivot_df = grouped_df.pivot(index="Quarter", columns="Year", values="Shipments")
sns.lineplot(data=pivot_df, markers=True,linewidth=2.7,markersize=8)
plt.title("Year-on-Year Quarterly Comparison of Shipments",fontsize=25)
plt.xlabel("Quarter",fontsize=17)
plt.ylabel("Shipments",fontsize=17)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
# Display the plot
plt.show()
```
